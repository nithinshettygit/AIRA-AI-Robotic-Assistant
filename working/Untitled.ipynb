```python
import time

# --- Dummy LLM Streaming Simulator ---
def fake_llm_with_markers_stream():
    lines = [
        "Hello students! Today we will learn something interesting.",
        "<<IMAGE_START:solar_system.png>>",
        "Here is an image showing the solar system with all the planets revolving around the sun. The planets are arranged according to their distance from the sun.",
        "<<IMAGE_END>>",
        "After this, let's discuss how gravity keeps planets in orbit and how the sun’s gravitational pull is the strongest in the solar system.",
        "<<IMAGE_START:rocket_launch.png>>",
        "This image shows a rocket launch, which uses powerful engines to escape Earth's gravity.",
        "<<IMAGE_END>>",
        "That concludes our lesson for today. Great job!"
    ]
    for line in lines:
        yield line
        time.sleep(1)  # simulate delay between narration lines


# --- Simulated real-time interface ---
print("\n--- STREAMING LESSON ---")
for chunk in fake_llm_with_markers_stream():
    if chunk.startswith("<<IMAGE_START"):
        image_file = chunk.split(":")[1].strip(">>")
        print(f"[UI] SHOW IMAGE: {image_file} (robot face disappears)")
    elif chunk.startswith("<<IMAGE_END"):
        print("[UI] SHOW ROBOT FACE AGAIN")
    elif chunk.strip():
        print(f"[TTS SPEAK]: {chunk}")
```
